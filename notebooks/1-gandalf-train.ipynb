# GANDALF train sample

```python
import sys
import os

# My modules
sys.path.append("../") # To support the use of the tool without packaging
from gandalf import GandalfTrain
```

## Arguments

```python
VERBOSE = 1
SUMMARY_MODELS = False
yaml_file = "train_config.yaml"
```

```python
gandalf_train = GandalfTrain.from_config_file(yaml_file)
```

```python
gandalf_train.show_arguments()
```

## Data

```python
dataset_train, dataset_test = gandalf_train.create_datasets(verbose=1)
```

## Models

```python
gandalf_train.create_autoencoder(summary_models=1)
```

```python
gandalf_train.create_discriminator(summary_models=1)
```

## Training

```python
gandalf_train.train(dataset_train, verbose=VERBOSE)
```

Evaluation

```python
ae_loss, reconstruction_loss, disc_loss = gandalf_train.eval_training(dataset_test)
```

## Save data

```python
gandalf_train.update_results_csv(reconstruction_loss, disc_loss, ae_loss)
```

```python
gandalf_train.save_visualization_data(dataset_train, dataset_test)
```
